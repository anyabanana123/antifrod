```python
from keras.models import Sequential
from keras.layers import Dense
import time

# Определение автокодировщика
autoencoder = Sequential()
autoencoder.add(Dense(30, input_dim=X_train_scaled.shape[1], activation='relu'))
autoencoder.add(Dense(15, activation='relu'))
autoencoder.add(Dense(30, activation='relu'))
autoencoder.add(Dense(X_train_scaled.shape[1], activation='sigmoid'))

# Компиляция модели
autoencoder.compile(optimizer='adam', loss='mse')

# Обучение модели
start_time = time.time()
autoencoder.fit(X_train_scaled, X_train_scaled, epochs=50, batch_size=256, shuffle=True, verbose=0)
training_time_autoencoder = time.time() - start_time
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Получение выходных данных автокодировщика для тестового набора
reconstructions = autoencoder.predict(X_test_scaled)

# Вычисление ошибки реконструкции (MSE)
mse = np.mean(np.power(X_test_scaled - reconstructions, 2), axis=1)

# Установка порога для классификации (например, 95-й процентиль ошибки реконструкции)
threshold = np.percentile(mse, 95)

# Классификация транзакций на основе порога
y_pred = (mse > threshold).astype(int)

# Оценка модели
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Autoencoder - Time: {training_time_autoencoder:.2f} seconds")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```
